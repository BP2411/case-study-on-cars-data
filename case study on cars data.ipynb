# Bar graph of male vs female buyers participated in the sales

```python
plt.title('Male vs Female Buyers')
fig=Data['Buyer Gender'].value_counts().plot(kind='bar', color=['red','blue'])
plt.xlabel('Gender')
plt.ylabel('Number of buyer')
plt.show()
```
